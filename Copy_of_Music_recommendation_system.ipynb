```python
df['Album/Movie']=df['Album/Movie'].str.replace(' ','')
df['Singer/Artists']=df['Singer/Artists'].str.replace(' ','')
df
```

```python
df['Singer/Artists']=df['Singer/Artists'].str.replace(',',' ')
df
```

```python
df['tags']=df['Singer/Artists']+' '+df['Genre']+' '+df['Album/Movie']+' '+df['User-Rating']
df['tags'][0]
df
```

```python
new_df=df[['Song-Name','tags']]
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2000)

# Láº¥y data cá»§a mÃ¬nh Ä‘á»ƒ lÃ m táº­p tá»« vá»±ng: 2000 lÃ  láº¥y 2000 tá»« phá»• biá»ƒn nháº¥t
```

```python
new_df.shape
new_df['tags'].shape
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
# vectors.shape
print(cv.get_feature_names_out().shape)  #  1697 : láº¥y Ä‘Æ°á»£c 1697 tá»« phá»• biáº¿n
print(cv.get_feature_names_out())  #  1697 : láº¥y Ä‘Æ°á»£c 1697 tá»« phá»• biáº¿n
print(vectors.shape) # 2391, 1697


# XÃ¢y dá»±ng tá»« vá»±ng: Giáº£ sá»­ táº¥t cáº£ cÃ¡c tá»« Ä‘á»u náº±m trong top 2000 tá»« phá»• biáº¿n nháº¥t, tá»« vá»±ng sáº½ lÃ :
#['activities', 'fun', 'is', 'love', 'programming'].

# Biáº¿n Ä‘á»•i vÄƒn báº£n thÃ nh vector sá»‘:

# "I love programming." -> [0, 0, 0, 1, 1]
# "Programming is fun." -> [0, 1, 1, 0, 1]
# "I love fun activities." -> [1, 1, 0, 1, 0]

# Há»‰u nuÃ´n
vectors[0,0:10]
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
print(similarity)
print(similarity.shape)

# ÄÃ¢y lÃ  ma tráº­n vuÃ´ng: HÃ ng gá»“m 2391 vector, coojtj cÅ©ng váº­y: Láº­t ngang dá»c ra áº¥y
# VÃ­ dá»¥: Sim[0,1]: hÃ ng 0 cá»™t 1: lÃ  0.22 ( ThÃ¬ pháº§n tá»­ á»Ÿ hÃ ng 0 lÃ  má»™t vector gá»“m (1 hÃ ng , 1697 cá»™t) so sÃ¡nh vá»›i ( 1 hÃ ng, 1697 cá»™t))
# á»Ÿ giÅ©a lÃ  1 háº¿t vÃ¬ so sÃ¡nh vá»›i chÃ­nh nÃ³
# oke
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[0])
```

```python
new_df.rename(columns={'Song-Name':'title'},inplace=True)
```

```python
def recommend(music):
    music_index=new_df[new_df['title']==music].index[0]
    distances=similarity[music_index]
    music_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in music_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Proper Patola')
```
