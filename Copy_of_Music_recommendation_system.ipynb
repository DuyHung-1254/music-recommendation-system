```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv(r"/content/ex.csv")
df
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df=df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
df.head()
```

```python
df['User-Rating']
```

```python
l=[]
for i in df['User-Rating']:
    l.append(i[:3])
```

```python
df.loc[:,'User-Rating'] = l
df
```

```python
df['Album/Movie']=df['Album/Movie'].str.replace(' ','')
df['Singer/Artists']=df['Singer/Artists'].str.replace(' ','')
df
```

```python
df['Singer/Artists']=df['Singer/Artists'].str.replace(',',' ')
df
```

```python
df['tags']=df['Singer/Artists']+' '+df['Genre']+' '+df['Album/Movie']+' '+df['User-Rating']
df['tags'][0]
df
```

```python
new_df=df[['Song-Name','tags']]
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2000)

# Lấy data của mình để làm tập từ vựng: 2000 là lấy 2000 từ phổ biển nhất
```

```python
new_df.shape
new_df['tags'].shape
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
# vectors.shape
print(cv.get_feature_names_out().shape)  #  1697 : lấy được 1697 từ phổ biến
print(cv.get_feature_names_out())  #  1697 : lấy được 1697 từ phổ biến
print(vectors.shape) # 2391, 1697


# Xây dựng từ vựng: Giả sử tất cả các từ đều nằm trong top 2000 từ phổ biến nhất, từ vựng sẽ là:
#['activities', 'fun', 'is', 'love', 'programming'].

# Biến đổi văn bản thành vector số:

# "I love programming." -> [0, 0, 0, 1, 1]
# "Programming is fun." -> [0, 1, 1, 0, 1]
# "I love fun activities." -> [1, 1, 0, 1, 0]

# Hỉu nuôn
vectors[0,0:10]
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
print(similarity)
print(similarity.shape)

# Đây là ma trận vuông: Hàng gồm 2391 vector, coojtj cũng vậy: Lật ngang dọc ra ấy
# Ví dụ: Sim[0,1]: hàng 0 cột 1: là 0.22 ( Thì phần tử ở hàng 0 là một vector gồm (1 hàng , 1697 cột) so sánh với ( 1 hàng, 1697 cột))
# ở giũa là 1 hết vì so sánh với chính nó
# oke
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[0])
```

```python
new_df.rename(columns={'Song-Name':'title'},inplace=True)
```

```python
def recommend(music):
    music_index=new_df[new_df['title']==music].index[0]
    distances=similarity[music_index]
    music_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in music_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Proper Patola')
```
